**3)**

(a) Show that $(n + 1)^5$ is $O(n^5)$.

(b) Show that $2^{n+1}$ is $O(2^n)$.

(c) Show that $n$ is $O(n \log n)$.

**Solution:**

(a) Since $\lim_{n \to \infty} \frac{(n+1)^5}{n^5}$ = $\lim_{n \to \infty} (\frac{n+1}{n})^5$ = $\lim_{n \to \infty} (1 + \frac{1}{n})^5$ = $1$.

Because $lim_{n \to \infty}\frac{f(n)}{g(n)}$ is constant, (n+1)^5 grows at the same rate as n^5. 

Therefore, $n^5$ is the theta of $(n+1)^5$ and then also the Big-O of $(n+1)^5$.

(b) Since $\lim_{n \to \infty} \frac{2^{n+1}}{2^n}$ = $\lim_{n \to \infty} 2$ = $2$

Because $lim_{n \to \infty}\frac{f(n)}{g(n)}$ is constant, 2^{n+1} grows at the same rate as $2^n$. 

Therefore, $2^n$ is the theta of $2^{n+1}$ and then also the Big-O of $2^{n+1}$.

(c) Since $\lim_{n \to \infty} \frac{n}{n \log n}$ = $\lim_{n \to \infty} \frac{1}{\log n}$ = $0$

Because $lim_{n \to \infty}\frac{f(n)}{g(n)}$ is 0, $n \log n$ grows faster than $n$. 

Therefore, $n \log n$ is the Big-O of $n$.

(b) The number of operations executed by algorithms A and B is $40n^2$ and $2n^3$, respectively. Determine $n_0$ such that A is better than B for $n \geq n_0$

**Solution:**

(a) A is better than B if number of operations executed by A is less than that of B.

Then $8n \log n$ $<$ $2n^2$.

Divide $2n$ from both sides; $4 \log n$ $<$ $n$.

Then we have $2^{4 \log n}$ $<$ $2^n$.

By the logarithm property, $2^{4 \log n}$ = $n^4$.

Now  $n^4$ $<$ $2^n$ is the condition we need to satisfy, which can be satisfied for all $n \geq 17$.

Therefore, $n_0 = 17$.

(b) A is better than B if number of operations executed by A is less than that of B.

Then $40n^2 < 2n^3$.

Divide $n^2$ from both sides; $40 < 2n$.

So we have $20 < n$ which can be satisfied by all $n \geq 21$.

Therefore, $n_0 = 21$.










**5)**
 Order the following 14 functions by growth rate.

$n, \sqrt n, n^{1.5}, n^2, n\log n, n \log \log n, n \log^2 n, n \log (n^2), 2/n, 2^n, 2^{n/2}, 37, n^2 \log n, n^3$

**Solution:**

We will classify the order of growth of each function first.

$n$ is linear.

$\sqrt n$ has lower growth rate compared to $n$ since $\lim_{n \to \infty}\frac{\sqrt n}{n} = \lim_{n \to \infty}\frac{1}{\sqrt n} = 0$.

$n^{1.5}$ has higher growth rate compared to $n$ since $\lim_{n \to \infty}\frac{n^{1.5}}{n} = \lim_{n \to \infty}\sqrt n$ which goes to $\infty$.

$n^2$ has higher growth rate compared to $n^{1.5}$ since $\lim_{n \to \infty}\frac{n^{2}}{n^{1.5}} = \lim_{n \to \infty}\sqrt n$ which goes to $\infty$.

So now we have $\sqrt n, n, n^{1.5}, n^2$ in increasing order of growth.

$n\log n$ has higher growth rate compared to $n$ since $\lim_{n \to \infty}\frac{n \log n}{n} = \lim_{n \to \infty}\log n$ which goes to $\infty$.

$n\log n$ has lower growth rate compared to $n^{1.5}$ since $\lim_{n \to \infty}\frac{n \log n}{n^{1.5}} = \lim_{n \to \infty}\frac{\log n}{\sqrt n} = 0$ since $\sqrt n$ grows much faster than $\log n$ when n is large(Imagine $2^{\sqrt n}$ which grows exponentially and $2^{\log n} = n$ which grows linearly).

$n\log \log n$ has higher growth rate compared to $n$ since $\lim_{n \to \infty}\frac{n \log \log n}{n} = \lim_{n \to \infty}\log \log n$ which goes to $\infty$.

$n\log \log n$ has lower growth rate compared to $n \log n$ since $\lim_{n \to \infty}\frac{n \log \log n}{n \log n} = \lim_{n \to \infty}\frac{\log \log n}{\log n} = 0$ since $\log n$ grows faster than $\log \log n$ (Imagine $2^{\log \log n} = \log n$ which grows logarithmically and $2^{\log n} = n$ which grows linearly, Logarithmically is slower than linear).

So now we have $\sqrt n, n, n \log \log n, n \log n, n^{1.5}, n^2$ in increasing order of growth.

$n \log^2 n$ has higher growth than $n \log n$  since $\lim_{n \to \infty}\frac{n \log^2 n}{n \log n} = \lim_{n \to \infty}\log n$ which goes to $\infty$.

$n \log^2 n$ has lower growth than $n^{1.5}$ since $\lim_{n \to \infty}\frac{n \log^2 n}{n^{1.5}} = \lim_{n \to \infty}\frac{\log^2 n}{\sqrt n} = 0$ (Imagine writen $n = 2^k$ for some real number $k$, we have $\log^2 n = k^2$ which is quadratic while $\sqrt n = n^{k/2}$ which is exponential).

$n \log (n^2) = 2n \log$ has the same growth rate as $n \log n$ since $\lim_{n \to \infty}\frac{2n \log n}{n \log n} =  \lim_{n \to \infty}2 = 2$ which is a constant.

So now we have $\sqrt n, n, n \log \log n, n \log n, n \log n^2, n \log^2 n, n^{1.5}, n^2$ in increasing order of growth.

$37$ is a constant, so it has lower growth than $\sqrt n$.

$2/n$ is decreasing when n increases, so it has lower growth than $37$.

So now we have $2/n, 37, \sqrt n, n, n \log \log n, n \log n, n \log n^2, n \log^2 n, n^{1.5}, n^2$ in increasing order of growth.

$n^2 \log n$ has higher growth rate than $n^2$ since $\lim_{n \to \infty}\frac{n^2 \log n}{n^2} = \lim_{n \to \infty}\log n$ which goes to $\infty$.

$n^3$ has higher growth rate than $n^2 \log n$ since  $\lim_{n \to \infty}\frac{n^3}{n^2 \log n} = \lim_{n \to \infty}\frac{n}{\log n}$ which goes to $\infty$ ($n$ grows faster than $\log n$).

$2^{n/2}$ grows exponentially, so it grows faster than $n^3$.

$2^n$ grows faster than $2^{n/2}$ since $\lim_{n \to \infty}\frac{2^n}{2^{n/2}} = \lim_{n \to \infty}2^{n/2}$ which goes to $\infty$.

Therefore, the order of function by increasing growth rate is $2/n, 37, \sqrt n, n, n \log \log n, n \log n, n \log n^2, n \log^2 n, n^{1.5}, n^2, n^2 \log n, n^3, 2^{n/2}, 2^n$.










**6)**

Suppose $T_1(n) = O(f(n))$ and $T_2(n) = O(f(n))$. Which of the following are always true (for all $T_1 , T_2$, and $f$)?

(a) $T_1 (n) + T_2 (n) = O(f(n))$

(b) $T_1 (n) − T_2 (n) = Θ(f(n))$

(c) $\frac{T_1(n)}{T_2(n)} = O(1)$

(d) $T_1 (n) = O(T_2 (n))$

**Solution:**

(a) True

(b) False

Think of $f(n) = n$.

Then $T_1(n) = O(n)$ and also $T_2(n) = O(n)$.

That means we can pick $T_1(n) = T_2(n) = n$

Therefore we have  $T_1(n) - T_2(n)$ = $n - n$ = $0$

If $T_1 (n) − T_2 (n) = Θ(n)$, there exists positive constant $c_1, c_2, and n_0$ such that 

$0 \leq c_1 * n \leq T_1 (n) − T_2 (n) \leq c_2 * n$ for $n \geq n_0$

Then we have $0 \leq c_1 * n \leq 0 \leq c_2 * n$ which is impossible.

Therefore (b) is false.

(c) False

Think of $f(n) = n^2$.

Then $T_1(n) = O(n^2)$ and also $T_2(n) = O(n^2)$.

That means we can pick $T_1(n) =  n^2$ but $T_2(n) = n$

Therefore $\frac{T_1(n)}{T_2(n)} = n$ which is not $O(1)$.

(d) False

Think of $f(n) = n^2$.

Then $T_1(n) = O(n^2)$ and also $T_2(n) = O(n^2)$.

That means we can pick $T_1(n) =  n^2$ but $T_2(n) = n$

Therefore $T_1 (n) =  n^2 \neq O(n) = O(T_2 (n))$
